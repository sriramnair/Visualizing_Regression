```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
d = sns.load_dataset('iris')
d.columns
```

```python
diction = {'sepal_length': d['sepal_length'],
           'sepal_width': d['sepal_width'],
           'petal_length': d['petal_length'],
           'petal_width': d['petal_width'],
           'species': d['species']}
data = pd.DataFrame(diction)
```

```python
data['is_setosa'] = data['species'].apply(lambda x: 1 if x=='setosa' else 0)
```

```python
new_dataFrame = data[['petal_length', 'is_setosa']]
```

```python
new_dataFrame
```

```python
sns.lmplot(data= new_dataFrame,x='petal_length', y='is_setosa')
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.loc[:,['sepal_length']], data.loc[:,['sepal_width']], test_size=.2)
```

```python
data
```

```python
sc = StandardScaler()
```

```python
X = data.loc[:,['sepal_length']]
X = sc.fit_transform(X_train)
```

```python
linear_regression = LinearRegression()
model = linear_regression.fit(X_train, y_train)
prediction = model.predict(X_test)
```
